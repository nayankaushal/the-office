```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests
import seaborn as sns
import re
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/54.0.2840.71 Safari/537.36'}
```

```python
ep_list = {1:6, 2:22, 3:23, 4:14, 5:26, 6:24, 7:24, 8:24, 9:23}
```

```python
a=[]
ep=[]
for i in range(1,10):
    for j in range(1, ep_list[i]+1):
        print('.', end='')
        if j<10:
            url = "https://www.officequotes.net/no{}-0{}.php".format(i,j)
        else:
            url = "https://www.officequotes.net/no{}-{}.php".format(i,j)
        r = requests.get(url, headers=headers)
        html = r.text
        soup = BeautifulSoup(html)
        x = soup.find_all(['a','div', 'span','h1'], class_ = ["quote","entry-title"])
        try:
            ep_title = x[0].text
        except:
            continue
        x.pop(0)
        for t in x:
            ep.append(ep_title)
            a.append(t.text)
value = a
```

```python
character =[]
dialogue = []
episode = []
b=" "
for i in range(0,len(value)):
        a_spl = value[i].split(":")
        for j in range(0, len(a_spl)-1):
            try:
                episode.append(ep[i])
                character.append(a_spl[j].split()[-1])
                dialogue.append(a_spl[j+1])
                if a_spl[j+1]!=a_spl[-1]:
                    dialogue[-1] = b.join(dialogue[-1].split()[:-1])
            except:
                continue
```

```python
dialogue = [dialogue[idx].strip("  \t\t\t\t\t") for idx in range(0, len(dialogue))]
```

```python
episode = pd.Series(episode)
character = pd.Series(character)
dialogue = pd.Series(dialogue)
```

```python
df = pd.concat([episode, character, dialogue], axis=1)
```

```python
df.columns = ['episode' ,'character', 'dialogue']
```

```python
df.to_csv("office_1.csv")
```
